# Bibliotecas necesarias

```python
from os import listdir
```

# Carga de categorías 


```python
DIR = 'corpus/AuthorProfilingTrack_Train/Author_profiling_track.train.truth'
consultas = dict()
with open(DIR,'r') as f:
  for doc in f.read().split('\n'):
      if doc:
          consultas[doc.split(':::')[0]] = {'lugar':doc.split(':::')[1],'profesion':doc.split(':::')[2]}
    
```

**Contenido del diccionario**

```python
print(consultas)
```

# Algunas estadísticas

```python
estadisticas = dict()

for name in consultas:
    lugar = estadisticas.get(consultas[name]['lugar'],0)
    profesion = estadisticas.get(consultas[name]['profesion'],0)
    estadisticas[consultas[name]['lugar']] = lugar + 1
    estadisticas[consultas[name]['profesion']] = profesion + 1
```

**Resultados**

```python
print(estadisticas)
```

# Carga del corpus


```python
DIR_CORPUS = 'corpus/AuthorProfilingTrack_Train/Train/'

ids = listdir(DIR_CORPUS)

textos = dict()

for name in ids:
    with open(DIR_CORPUS + name,'r') as f:
        textos[name] = f.read()
```

**Contenido del diccionario**
